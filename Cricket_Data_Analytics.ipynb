## Match Results data file

```python
url= 'https://raw.githubusercontent.com/arghanath007/Cricket-Data-Analytics/master/Cricket%20Data/T20_world_cup_match_results.json'
json_texts= requests.get(url).text
json_data= json.loads(json_texts)
json_data
```

```python
len(json_data)
```

```python
json_data[0]
```

### Getting all of the match results

```python
json_data[0]['matchSummary']
```

### Converting into DataFrame

```python
df_match_summary =pd.DataFrame(json_data[0]['matchSummary'])
df_match_summary.head()
```

```python
len(df_match_summary)
```

```python
df_match_summary.shape
```

### Renaming column name

```python
df_match_summary.rename({'scorecard': 'match_id'}, axis=1, inplace=True)
df_match_summary
```

### Converting the `match_summary` dataframe into a csv file.

```python
df_match_summary.to_csv('T20_match_Data/match_summary.csv', index=False)
```

### Creating a dictionary
> For enabling cross communication between two or more tables.

image.png

```python
match_ids_dict = {}

for index, row in df_match_summary.iterrows():
  key1 = row['team1'] + ' Vs ' + row['team2']
  key2 = row['team2'] + ' Vs ' + row['team1']

  match_ids_dict[key1]= row['match_id']
  match_ids_dict[key2]= row['match_id']

match_ids_dict
```

### Transforming the `dismissal` column into `Out/Not out` column

> We can see that if there are `some string` in the `dismissal` column then that player/cricketer is `out`, otherwise the player/cricketer is `not out`.

### Was getting an error

```python
# df_batting_summary['Out/Not out']= df_batting_summary.dismissal.apply(lambda x: 'Out' if len(x) > 0 else 'Not out')
# df_batting_summary

df_batting_summary['Out/Not out'] = df_batting_summary.dismissal.apply(lambda x: 'Out' if isinstance(x, (str, list)) and len(x) > 0 else 'Not out')
df_batting_summary
```

### Dropping the `dismissal` column

```python
df_batting_summary.drop('dismissal', axis=1, inplace=True)
df_batting_summary.head()
```

### Before

```python
df_batting_summary['batsmanName'].head(20)
```

### After

```python
df_batting_summary['batsmanName'] = df_batting_summary['batsmanName'].apply(lambda s: ''.join(re.findall(r'[\w]+', s)))
df_batting_summary['batsmanName'].head(20)
```

```python
df_batting_summary.head(20)
```

### Creating a `match_id` column

> For creating connections between two tables. It will be useful in `PowerBi`.

```python
match_ids_dict['Namibia Vs Sri Lanka']
```

```python
match_ids_dict['U.A.E. Vs Netherlands']
```

```python
df_batting_summary['match_id'] = df_batting_summary['match'].map(match_ids_dict)

df_batting_summary.head(25)
```

### Converting the `batting_summary` dataframe into a csv file

```python
df_batting_summary.to_csv('T20_match_Data/batting_summary.csv', index=False)
```
